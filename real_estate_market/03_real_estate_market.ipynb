# Исследование объявлений о продаже квартир

Необходимо определить рыночную стоимость объектов недвижимости с помощью данных сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**План работы**

1. Изучение и предобработка данных: 
- проверка на дубликаты; 
- пропуски; 
- замена пропусков и типов данных там, где это уместно.

2. Проведение исследовательского анализа:
- изучение площади, цены, числа комнат, высоты потолков;
- изучение времени продажи квартиры; 
- очистка данных от редких и выбивающихся значений;
- изучение зависимости цены от различных факторов — площади, числа комнат, удалённости от центра, на каком этаже расположена квартира, от даты размещения объявления;
- определение топ-10 населённых пунктов с наибольшим числом объявлений, расчет средней цены квадратного метра в этих населённых пунктах;
- определение области, входящей в центр Санкт-Петербурга;
- анализ квартир в центре по площади, цене, числу комнат, высоте потолков;
- определение факторов, влияющих на стоимость квартиры в центре. 

**В результате работы мы ответим на вопросы:**
1. Сколько обычно занимает продажа. Когда можно считать, что продажи прошли очень быстро, а когда необычно долго?
2. В каких населённых пунктах самая высокая и низкая стоимость жилья?
3. Какие факторы больше всего влияют на стоимость квартиры? 
4. Какие факторы влияют на стоимость квартир в центре Санкт-Петербурга? Отличаются ли они от факторов для квартир из других областей города или населенных пунктов?

Рассчитаем стоимость квадратного метра и добавим столбец с этим значением в таблицу.

```python
df.insert(3, 'price_square_meter', 0) # создаем новый столбец рядом с 'total_area'
df['price_square_meter'] = (df['last_price'] / df['total_area']).astype('int') # рассчитываем стоимость
df = df.reset_index()
display(df.head()) # проверяем результат
```

### Дата публикации

### Этаж

Проведем категоризацию по этажам. Напишем функцию, которая разделит данные на основании информации об этаже по трем категориям: "первый", "последний" и "другой".

```python
def floor_category(row): # задаем имя функции и аргумент
    floors_total = row['floors_total'] # задаем переменную для последнего этажа
    floor = row['floor'] # задаем переменную для фактического этажа
    if floor == 1: # определяем условия для каждой категории
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'

df.insert(10, 'floor_category', 0) # создаем новый столбец рядом с 'floor'
df['floor_category'] = df.apply(floor_category, axis = 1) # добавляем в созданный столбец
display(df[['floor', 'floor_category']].head()) # проверим результат
```

### Соотношение площадей

Посчитаем соотношение жилой и общей площади, а также отношение площади кухни к общей и добавим эту информацию в таблицу.

```python
df.insert(9, 'useful_area_ratio', 0) # создаем новый столбец рядом с 'living_area'
df['useful_area_ratio'] = round((df['living_area'] / df['total_area']), 1) # рассчитываем соотношение, сокращаем до 1 знака после запятой
df.insert(10, 'kitchen_area_ratio', 0) # создаем новый столбец рядом с 'useful_area_ratio'
df['kitchen_area_ratio'] = round((df['kitchen_area'] / df['total_area']), 1) # рассчитываем соотношение, сокращаем до 1 знака после запятой
display(df[['living_area', 'useful_area_ratio', 'kitchen_area_ratio']].head()) # проверим результат
```

#### Удаленность от центра

Чтобы облегчить работу, категоризируем удаленность квартиры от центра с помощью функции.

```python
def flats_category(row): # задаем имя функции и аргумент
    if row['cityCenters_nearest'] < 2500: # определяем условия для каждой категории
        return 'center'
    elif row['cityCenters_nearest'] < 5000:
        return 'near_center'
    elif row['cityCenters_nearest'] < 17000:
        return 'in_the_city'
    elif row['cityCenters_nearest'] >= 17000:
        return 'outside_the_city'
df['flats_category'] = df.apply(flats_category, axis=1) # создаем новый столбец и добавляем в него значения
df[['locality_name', 'cityCenters_nearest', 'flats_category']].head() # проверим результат
```

Теперь сделаем диаграмму:

```python
pivot_table_locality = df.pivot_table(
    index='flats_category', 
    values='last_price', 
    aggfunc=['mean', 'count', 'median']
).reset_index()
pivot_table_locality.columns = ['flats_category', 'mean', 'count', 'median'] # убираем мультииндекс
# строим диаграмму
pivot_table_locality.plot(x='flats_category', 
                          y='median', 
                          kind='bar', 
                          title='Зависимость стоимости от расположения', 
                          figsize=(9,4)
)
plt.show()
```

Квартиры в центре – самые дорогие, затем, немного ниже по стоимости квартиры недалеко от центра, гораздо дешевле квартиры на окраинах города и в области.

#### Этаж
